```python
#import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import re

#import list of files

directory=r'E:\CAREER_DATA\data_files\data_visualization/*'
directory=glob.glob(directory)
list_of_files=[]

for i in directory:
    # Regular expression pattern to extract the file name
    pattern = r'\\([^\\]+)$'
    
    # Search for the pattern in the string
    match = re.search(pattern, i)

    if match:
        file_name_with_extension = match.group(1)
        list_of_files.append(file_name_with_extension)
        print("Extracted file name:", file_name_with_extension)
    else:
        print("File name not found.")
        
print("USE file_dir TO ACCESS DIRECTORY")
file_dir="E:\CAREER_DATA\data_files\data_visualization\\"
#file=input("Enter File Name:")
#achieved_file=pd.read_csv(file_dir+file,index_col=0)
```

<h1> Line Plot</h1>
<h3>When do we use it</h3>
<p>Line plots are use to show relationship between <b>two NUMERICAL CONTINEOUS LINEAR variable.</b><br>
    What type of data can be used</p>
    <ul>
    <li>Time series data: where you have data points collected at regular intervals over time. This can be used to analyze trends, seasonal patterns, or changes over time.</li>
    <li>Continuous data: temperature readings, stock prices, or population growth, where the values can take any numeric value within a range</li>
    <li>Comparing trends:When you have multiple lines on the same plot, you can easily compare trends and spot any patterns or correlations between the variables</li>
    <li>To validate data continuity:useful when you want to emphasize the progression or lack of gaps in the data</li>
    <li>Identifying outliers:Line plots can help identify outliers in the data, which are data points that deviate significantly from the overall trend</li>
    <li>Forecasting and predictions: Line plots can be used to visualize historical data and make predictions or projections based on trends</li>
    </ul>
    <p><b>IMPORTANT MESSAGE</b><br>
    Remember, the effectiveness of a line plot depends on the type of data and the story you want to convey. Always consider the nature of your data and your specific analysis goals when choosing the appropriate type of visualization.
    </p>

```python
#decide which file to import
#criteria to import file is to whether date time column exist or not, or atleast date column exist that later converted to datetime column
%who
for i in list_of_files:
    print(i)
#print(file_dir)
summer_df=pd.read_csv(file_dir+"summer2016.csv")
#summer_df.info() #it doesn't have the date column
print(seattle_weather_df.info())
print(seattle_weather_df["DATE"])
print(seattle_weather_df)
```

```python
#we will be using seattle weather and convert its date column to datetime column
#first understand that if a date column has int64 which is repating from 1-12 then it is likely about a month
#in this case we have to assume dates and convert it to datetime column

day_seattle=str(1)
year_seattle=str(2007)
seattle_weather_df["MONTH"]=pd.to_datetime(year_seattle + "-" + seattle_weather_df["DATE"].astype(str) + "-"+ day_seattle,
                                        format="%Y-%m-%d")
#creating a month column and set the dates:
#convert column to str
#concatenating in date format with year and day
#defining format by format="%m-%Y-%d"
seattle_weather_df["MONTH"]
#datetime column has created
```

```python
#plotting timeseries data

if seattle_weather_df.index.name is None: #it check whether the index name set to None? if true then set index as MONTH
    seattle_weather_df.set_index("MONTH", inplace=True)
#MONTH column set as index of dataframe, inplace argument is insuring the changes appear in original dataframe, but running twice will create an error
#print(seattle_weather_df.index) #converted column to index

#now plot
plt.plot(seattle_weather_df.index, seattle_weather_df["MLY-CLDD-BASE45"])
plt.show()
```
