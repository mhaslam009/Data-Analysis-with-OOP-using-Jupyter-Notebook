<h1>What DATA is suitable for LINE PLOT</h1>
<p><b>VALUES MUST BE LINEAR AND CONTINOUS not repeating</b><br>
    Type of Data:</p>
<ul>
    <li>Time Series Data: monthly temperature readings, daily stock prices, hourly website traffic</li>
    <li>Continuous Function or Curve: a sine wave, exponential growth, or any other mathematical function</li>
    <li>Continuous Measurement Data: population growth, sales over time, or sensor reading</li>
    </ul>

```python
#Since above graph values are not sorted we wll sort both columns
line_plot_sort_two_columns=seattle_weather_df.sort_values(by=["DATE"],
                                                          ascending=False)
plt.plot(line_plot_sort_two_columns["DATE"],line_plot_sort_two_columns["MLY-CLDD-BASE45"])
plt.show()
plt.scatter(line_plot_sort_two_columns["DATE"],line_plot_sort_two_columns["MLY-CLDD-BASE45"])
plt.show()
```

```python
austin_weather_df
#line plot with customization
fig, axes=plt.subplots(nrows=2,ncols=2,figsize=(10,10),sharey=True)
#1st argument is number of rows layout for graphs
#2nd argument is number of cols layout for graphs
#3rd argument is each figure size, left one is increasing==increasing width, right one increasing==increasing height
#4th argument is making each graph y axis scaling simmiler
axes[0,0].plot(austin_weather_df["DATE"],austin_weather_df["MLY-CLDD-BASE45"])
#with marker,linestyle and color
axes[0,1].plot(austin_weather_df["DATE"],austin_weather_df["MLY-CLDD-BASE45"],
              marker="v",linestyle="--",color="b")
axes[1,0].plot(austin_weather_df["DATE"],austin_weather_df["MLY-CLDD-BASE45"],
              marker="o",linestyle="--",color="r")
axes[1,1].plot(austin_weather_df["DATE"],austin_weather_df["MLY-CLDD-BASE45"],
              marker="o",linestyle="None",color="r")
#to customize the labels(x or y)
axes[1,1].plot(austin_weather_df["DATE"],austin_weather_df["MLY-CLDD-BASE45"],
                 marker="o")
axes[1,1].set_xlabel("Time")
axes[1,1].set_ylabel("Temperature")
axes[1,1].set_title("Austin weather station#45")

plt.subplots_adjust(wspace=0.6, hspace=0.6) #adjust spacing between graphs
#1st argument increase==increase space between graphs in x-axis
#2nd argument increase==increase space between graphs in y-axis

fig.tight_layout(w_pad=0.5, h_pad=0.8) #increase or decrease each figure size
#1st argument increase==increase each graph in width
#2nd argument increase==increase each graph in height

#now to manage the size of title of graphs
axes[1,1].set_title("Austin Temperature",fontsize=5)
axes[1,1].set_xlabel("Time",fontsize=20)
axes[1,1].set_ylabel("Temperature",fontsize=10)
plt.show()
```

```python
#plotting time series

climate_change_df=pd.read_csv(
    "E:/CAREER_DATA/data_files/data_visualization/climate_change.csv", parse_dates=["date"],index_col="date")
#2nd argument is converting "date" column from object type to datetime object
#3rd argument is set new index to "date"
#why because we are doing time series plotting, for that index needed to be set as dates

climate_change_df

fig_time,axes_time=plt.subplots(nrows=2,ncols=2,figsize=(10,10),sharey=True)
axes_time[0,0].plot(climate_change_df.index,climate_change_df["co2"])
#print(climate_change_df.columns)


#how can you zoom in(subset years) 
#subset for 10 years
subset_10_years=climate_change_df["1960-01-01":"1969-12-31"]
#plot it
axes_time[0,1].plot(subset_10_years.index,subset_10_years["co2"])
#as you know the x_labels are become tidy so we rotate it
axes_time[0,1].tick_params(axis='x',rotation=45)


#subset 1 year
subset_a_year=climate_change_df["1960-01-01":"1960-12-31"]
#plot it
axes_time[1,0].plot(subset_a_year.index,subset_a_year["co2"],
                    marker="o",color="r")
#rotate it
axes_time[1,0].tick_params(axis="x",rotation=45)
plt.show()


```

```python
#annotate plots
fig_annotate, axes_annotate=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
axes_annotate[0].plot(climate_change_df.index, climate_change_df["relative_temp"])
axes_annotate[1].annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2),
             arrowprops={'arrowstyle':'->', 'color':'gray'})
#1st argument is the string you are putting on a graph
plt.show()

```
