```python
%who
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': ['a', 'b', 'c']})
df2 = pd.DataFrame({'A': [4, 5, 6], 'B': ['d', 'e', 'f']})
df3 = pd.DataFrame({'A': [7, 8, 9], 'B': ['g', 'h', 'i']})
print(df1)
print("\n")
print(df2)
print("\n")
print(df3)
print("\n")
#inner join
ij1=df1.merge(df2,on="A",suffixes=("_df1","_df2"))
#left join
lj2=df1.merge(df2,on="A",how="left",suffixes=("_df1","_df2"))
#right join
rj3=df1.merge(df2,on="A",how="right",suffixes=("_df1","_df2"))
#outer join
oj4=df1.merge(df2,on="A",how="outer",suffixes=("_df1","_df2"))
#self join
sj5=df1.merge(df1,on="A",suffixes=("_df1","_df2"))
#semi join
a=df1.merge(df2,on="A",)
semi_join=df1[df1["A"].isin(a["A"])]
#anti join
#x=df1.merge(df2,on="A",how="left")
#y=x[x["_merge"]=="left_only","gid"]
#anti_join=df1[df1["gid"].isin(y)]
print(ij1)
print("\n")
print(lj2)
print("\n")
print(rj3)
print("\n")
print(oj4)
print("\n")
print(sj5)
print("\n")
print(semi_join)
#print("\n")
#print(anti_join)
```

<h1>Result</h1>
<ul>
    <li>inner join: shows no content kyun keh joining table main kuch bhi common nahi tha elements main</li>
    <li>left join: B_df1 data dikha raha hai aur B_df2 main koi data nahi iska matlb common+left table ke elements hain bas</li>
    <li>right join: bilkul left join ke opposite operation</li>
    <li>outer join: saarey table ko combine krdya vertically</li>
    <li>self join: 1 table ko apas main combine kara tha toh naya table main srf df1 ka data ha</li>
    <li>semi-join: returns contents from left table including the common elements from both tables</li>
    <li>anti-join: returns contents from left table excluding common elements from both table</li>
    </ul>
