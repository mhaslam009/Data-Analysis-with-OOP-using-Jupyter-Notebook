```python
import glob
import pandas as pd

# Take the directory path as input from the user
directory_path = input("Enter the directory path: ")
directory_path += '/*'

# Get a list of all files in the directory
file_list = glob.glob(directory_path)

# Create empty dictionaries for dataframes
list_csv_df = {}
list_txt_df = {}

# Loop through each file and load it into a DataFrame
for file_path in file_list:
    if file_path.endswith(".csv"):
        # Extract the file name (without extension) from the file path
        file_name = file_path.split("/")[-1].split(".")[0]
        df_csv = pd.read_csv(file_path)
        list_csv_df[file_name] = df_csv

    elif file_path.endswith(".txt"):
        # Extract the file name (without extension) from the file path
        file_name = file_path.split("/")[-1].split(".")[0]
        df_txt = pd.read_csv(file_path)
        list_txt_df[file_name] = df_txt

# ... (The rest of your code remains the same)

# Take the file name as input from the user
file_name = input("Enter the file name: ")

# Check if the file name exists in the dictionaries
if file_name in list_csv_df:
    df = list_csv_df[file_name]
    print("DataFrame for file:", file_name)
    print(df)
elif file_name in list_txt_df:
    df = list_txt_df[file_name]
    print("DataFrame for file:", file_name)
    print(df)
else:
    print("File not found.")
```
